```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Import data and win predictions from linear regression model
pred= pd.read_csv("Data/predictions.csv")
data = pd.read_csv("Data/2000 to 2020 data.csv")
data_21 = pd.read_csv("Data/2021.csv")


pred.head()
```

```python
# Add conference to pred df

pred_conf = pd.merge(pred,  
                     data_21,  
                     on ='Team',  
                     how ='left') 
pred_conf
```

```python
# Divide into West and East conferences

is_west =  pred_conf['Conf']=='W'
is_east = pred_conf['Conf']=='E'

east_conf = pred_conf[is_east]


west_conf = pred_conf[is_west]
# west_conf
east_conf
```

```python
# Choose top 8 from each conference to qualify for playoffs

east_qual = east_conf.nlargest(8,'Predicted Wins')

east_qual

west_qual = west_conf.nlargest(8,'Predicted Wins')
west_qual
east_qual
```

```python
# Set brackets for east round 1

east_game1 = east_qual.iloc[[0,7], [0,1]] 
east_game1.reset_index(drop=True,inplace=True)
east_game1['Chance of Qualifying'] = (east_game1['Predicted Wins'] / east_game1['Predicted Wins'].sum()) * 100

east_game1_winner = east_game1.nlargest(1, 'Chance of Qualifying')


east_game2 = east_qual.iloc[[3,4], [0,1]] 
east_game2['Chance of Qualifying'] = (east_game2['Predicted Wins'] / east_game2['Predicted Wins'].sum()) * 100
east_game2.reset_index(drop=True,inplace=True)
east_game2_winner = east_game2.nlargest(1, 'Chance of Qualifying')


east_game3 = east_qual.iloc[[2,5], [0,1]] 
east_game3['Chance of Qualifying'] = (east_game3['Predicted Wins'] / east_game3['Predicted Wins'].sum()) * 100
east_game3.reset_index(drop=True,inplace=True)
east_game3_winner = east_game3.nlargest(1, 'Chance of Qualifying')

east_game4 = east_qual.iloc[[1,6], [0,1]] 
east_game4['Chance of Qualifying'] = (east_game4['Predicted Wins'] / east_game4['Predicted Wins'].sum()) * 100
east_game4.reset_index(drop=True,inplace=True)

east_game4_winner = east_game4.nlargest(1, 'Chance of Qualifying')


print('East Conference Playoffs Round 1 Game 1: ' + east_game1['Team'][0] + ' vs. ' + east_game1['Team'][1] )
print('East Conference Playoffs Round 1 Game 2: ' + east_game2['Team'][0] + ' vs. ' + east_game2['Team'][1] )
print('East Conference Playoffs Round 1 Game 3: ' + east_game3['Team'][0] + ' vs. ' + east_game3['Team'][1] )
print('East Conference Playoffs Round 1 Game 4: ' + east_game4['Team'][0] + ' vs. ' + east_game4['Team'][1] )
```

```python
# Set brackets for west round 1

west_game1 = west_qual.iloc[[0,7], [0,1]] 
west_game1['Chance of Qualifying'] = (west_game1['Predicted Wins'] / west_game1['Predicted Wins'].sum()) * 100
west_game1.reset_index(drop=True,inplace=True)
west_game1_winner = west_game1.nlargest(1, 'Chance of Qualifying')

west_game2 = west_qual.iloc[[3,4], [0,1]] 
west_game2['Chance of Qualifying'] = (west_game2['Predicted Wins'] / west_game2['Predicted Wins'].sum()) * 100
west_game2.reset_index(drop=True,inplace=True)
west_game2_winner = west_game2.nlargest(1, 'Chance of Qualifying')


west_game3 = west_qual.iloc[[2,5], [0,1]] 
west_game3['Chance of Qualifying'] = (west_game3['Predicted Wins'] / west_game3['Predicted Wins'].sum()) * 100
west_game3.reset_index(drop=True,inplace=True)
west_game3_winner = west_game3.nlargest(1, 'Chance of Qualifying')

west_game4 = west_qual.iloc[[1,6], [0,1]] 
west_game4['Chance of Qualifying'] = (west_game4['Predicted Wins'] / west_game4['Predicted Wins'].sum()) * 100
west_game4.reset_index(drop=True,inplace=True)

west_game4_winner = west_game4.nlargest(1, 'Chance of Qualifying')

print('West Conference Playoffs Round 1 Game 1: ' + west_game1['Team'][0] + ' vs. ' + west_game1['Team'][1] )
print('West Conference Playoffs Round 1 Game 2: ' + west_game2['Team'][0] + ' vs. ' + west_game2['Team'][1] )
print('West Conference Playoffs Round 1 Game 3: ' + west_game3['Team'][0] + ' vs. ' + west_game3['Team'][1] )
print('West Conference Playoffs Round 1 Game 4: ' + west_game4['Team'][0] + ' vs. ' + west_game4['Team'][1] )
```

```python
# Set brackets for east round 2

east_game5 = east_game1_winner.append(east_game2_winner)
east_game5['Chance of Qualifying'] = (east_game5['Predicted Wins'] / east_game5['Predicted Wins'].sum()) * 100
east_game5.reset_index(drop=True,inplace=True)
east_game5_winner = east_game5.nlargest(1, 'Chance of Qualifying')


east_game6 = east_game3_winner.append(east_game4_winner)
east_game6['Chance of Qualifying'] = (east_game6['Predicted Wins'] / east_game6['Predicted Wins'].sum()) * 100
east_game6.reset_index(drop=True,inplace=True)
east_game6_winner = east_game6.nlargest(1, 'Chance of Qualifying')


print('East Conference Playoffs Round 2 Game 1: ' + east_game5['Team'][0] + ' vs. ' + east_game5['Team'][1] )
print('East Conference Playoffs Round 2 Game 2: ' + east_game6['Team'][0] + ' vs. ' + east_game6['Team'][1] )
```

```python
# Set brackets for west round 2

west_game5 = west_game1_winner.append(west_game2_winner)
west_game5['Chance of Qualifying'] = (west_game5['Predicted Wins'] / west_game5['Predicted Wins'].sum()) * 100
west_game5.reset_index(drop=True,inplace=True)
west_game5_winner = west_game5.nlargest(1, 'Chance of Qualifying')


west_game6 = west_game3_winner.append(east_game4_winner)
west_game6['Chance of Qualifying'] = (east_game6['Predicted Wins'] / east_game6['Predicted Wins'].sum()) * 100
west_game6.reset_index(drop=True,inplace=True)
west_game6_winner = west_game6.nlargest(1, 'Chance of Qualifying')


print('West Conference Playoffs Round 2 Game 1: ' + west_game5['Team'][0] + ' vs. ' + west_game5['Team'][1] )
print('West Conference Playoffs Round 2 Game 2: ' + west_game6['Team'][0] + ' vs. ' + west_game6['Team'][1] )
```

```python
# Conference Finals

east_conf_final = east_game5_winner.append(east_game6_winner)
east_conf_final['Chance of Qualifying'] = (east_conf_final['Predicted Wins'] / east_conf_final['Predicted Wins'].sum()) * 100
east_conf_final.reset_index(drop=True,inplace=True)
east_conf_final_winner = east_conf_final.nlargest(1, 'Chance of Qualifying')

west_conf_final = west_game5_winner.append(west_game6_winner)
west_conf_final['Chance of Qualifying'] = (west_conf_final['Predicted Wins'] / west_conf_final['Predicted Wins'].sum()) * 100
west_conf_final.reset_index(drop=True,inplace=True)
west_conf_final_winner = west_conf_final.nlargest(1, 'Chance of Qualifying')

print('West Conference Finals: ' + east_conf_final['Team'][0] + ' vs. ' + east_conf_final['Team'][1] )
print('East Conference Finals: ' + west_conf_final['Team'][0] + ' vs. ' + west_conf_final['Team'][1] )


```

```python
# NBA Finals

nba_finals = east_conf_final_winner.append(west_conf_final_winner)
nba_finals['Chance of Qualifying'] = (nba_finals['Predicted Wins'] / nba_finals['Predicted Wins'].sum()) * 100
nba_finals.reset_index(drop=True,inplace=True)

nba_finals_champion = nba_finals.nlargest(1, 'Chance of Qualifying')
nba_finals_champion.reset_index(drop=True,inplace=True)
print('NBA Finals: ' + nba_finals['Team'][0] + ' vs. ' + nba_finals['Team'][1] )
print('---------------------------------------------------------------------')
print('NBA 2021  Projected Champion: ' + nba_finals_champion['Team'][0] + '!' )
```
